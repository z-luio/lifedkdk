```python
import matplotlib.pyplot as plt
import numpy as np
n = int(input("Enter geater  than 3 number : "))

def Prime(n):
    counter = 3
    prime = [2,3]
    if n % 2 == 0 :
        n -= 1
    while counter <n :
        counter +=2
        c = counter % np.array(prime)
        if False == (0 in c) :
            prime.append(counter)
    print(len(prime))
    return prime
        
        
        
    
    
y = np.array(Prime(n))
x = np.arange(1,y.size + 1)

# scatter
fig, ax = plt.subplots()

ax.scatter(x, y,label = 'prime' ,marker = '.' ,c ='#4c8d9c')
ax.set_title('prime',c = '#4c8d9c')
ax.legend()


plt.show()
```
